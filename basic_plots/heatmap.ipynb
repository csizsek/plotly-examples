```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

np.random.seed(1337)
```

```python
# Basic heatmap with a color bar

fig = go.Figure()
fig.add_trace(
    go.Heatmap(
        z=np.random.randn(30, 30), colorscale="Blues", colorbar={"title": "Legend"}
    )
)
fig.update_traces(hovertemplate="x: %{x}" + "<br>y: %{y}" + "<br>value: %{z:0.3f}")
fig.update_layout(template="plotly_white", width=800, height=800, title="Plot title")
fig.update_xaxes(visible=False)
fig.update_yaxes(visible=False)
fig.show()
```
