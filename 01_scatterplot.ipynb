# Scatterplot

```python
df = px.data.gapminder()
```

```python
fig = px.scatter(df[df.year == 2007],
                 x='gdpPercap',
                 y='lifeExp',
                 hover_data=['country'],
                 trendline='lowess',#'ols'
                 width=1100,
                 height=800)

fig.update_xaxes(range=[0, 50000], dtick=5000, showticklabels=True)
fig.update_yaxes(range=[0, 100], dtick=10, showticklabels=True)

fig.update_layout(xaxis_title='GDP per capita',
                  yaxis_title='Life expectancy')

fig.show()
```

```python
df['foo'] = np.random.randn(len(df))
fig = px.scatter(df[df.year == 2007],
                 x='gdpPercap',
                 y='lifeExp',
                 size='pop',
                 size_max=60,
                 hover_name='country',
                 hover_data=['iso_alpha'],
                 color='continent',
                 opacity=0.8,
                 log_x=True,
                 width=1100,
                 height=800)

fig.update_xaxes(range=[2, 5], tickangle=45, nticks=30, showticklabels=True)
fig.update_yaxes(range=[30, 90], tickangle=45, nticks=50, dtick=5, showticklabels=True)

fig.update_layout(title='Plot Title',
                  xaxis_title='GDP per capita',
                  yaxis_title='Life expectancy',
                  showlegend=True,
                  legend=dict(x=0.9, y=0),
                  legend_title=None,
                  #legend_orientation='h',
                  dragmode='pan',
                  xaxis_showgrid=False,
                  yaxis_showgrid=False)

fig.update_traces(marker=dict(line=dict(width=0.5,
                                        color='DarkSlateGrey')))

fig.show(config={'scrollZoom': True})
```
